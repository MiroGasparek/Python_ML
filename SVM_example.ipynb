# Support Vector Machine example
Here we will consider the use of the support vector machine. We generate the random data below and above the line to illustrate the suitable example for soft-margin Support Vector Machine use.

```python
def y_line(x, a=1, b=1):
    """ Define the y-cordinate of the linear function."""
    return a*x+b
```

```python
# Number of points
num_points = 10

# Set up the linear coefficients for the eq. of the line
a = 10
b = 5

# Set up parameters for the Gaussian
mean = 30
std = 40
# Generate the Numpy array with all the x-axis values
X = np.linspace(0,10,num_points)

# Generate the Numpy array for the line
X_line = np.linspace(-1,11,num_points)

# Generate the noisy y-values below and above the line
y_above = [y_line(x, a, b) + abs(random.gauss(mean,std)) for x in X]
y_below = [y_line(x, a, b) - abs(random.gauss(mean,std)) for x in X]

### Plotting 
fig  = plt.figure(figsize=(6, 4),dpi=100)

# Plot the dividing line
# y_true = [y_line(x, a, b) for x in X_line]

# Make a scatter plot of all the points and the line
plt.scatter(X, y_above, c='blue')
plt.scatter(X, y_below, c='red')
# plt.plot(X_line, y_true, c='black')

# Set labels
plt.xlabel(r'$\mathrm{x_{1}}$',fontsize=18)
plt.ylabel(r'$\mathrm{x_{2}}$',fontsize=18)
plt.title('Noisy data: suitable for soft-margin SVM classification', fontsize=16)
plt.legend(['data class 1', 'data class 2'])

# Set x-axis limits
# plt.xlim([-1, 11])

plt.show(fig)

# Save the figure
fig.savefig('SoftSVM_example.png')


```
